```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
ROOT_DIR = 'net/tagged_imgs_org/'

img = 255 - cv2.imread(os.path.join(ROOT_DIR, '00/00020.png'), cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
sumC = np.sum(img, axis=0) # 57
sumR = np.sum(img, axis=1) # 19

col = [i for i, val in enumerate(sumC) if val > 48]
row = [i for i, val in enumerate(sumR) if val > 48]
minX, maxX = min(col), max(col)+1
minY, maxY = min(row), max(row)+1
print(minX, maxX)
```

```python
output_img = np.zeros(img.shape)
nx, ny = maxX-minX, maxY-minY
sx, sy = (img.shape[1] - nx) // 2, (img.shape[0] - ny) // 2
output_img = cv2.resize(img[minY:maxY, minX:maxX], (57, 57))
# output_img[sy: sy+ny, sx: sx+nx] = img[minY:maxY, minX:maxX]

plt.imshow(output_img, cmap='gray')
```
